```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd                                #can't jupyter give this by default
```

```python
# Importing the dataset
dataset = pd.read_csv('Iris.csv')                                #data data all over the place!
X = dataset.iloc[:, [1,2,3]].values                              #attributes
y = dataset.iloc[:, -1].values                                   #bull's eye

dataset                                            #uh oh!! phuckin categorical values again :(
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset['Species']= label_encoder.fit_transform(dataset['Species'])
dataset
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)                           #fits perfectly, just like me in a group of extroverts :((
```

```python
from sklearn.naive_bayes import GaussianNB
clas = GaussianNB()
clas.fit(X_train, y_train)
```

```python
y_pred = clas.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
ac = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test, y_pred)
print(ac)
print(cm)
```
